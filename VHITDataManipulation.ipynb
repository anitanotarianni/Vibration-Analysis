```python
import pandas as pd

# Caricamento dei file Excel (assicurati che siano stati caricati in Colab o presenti nella directory)
main_iso = pd.read_excel('MainISOfiltriaperti.xlsx')
mod_50_5000 = pd.read_excel('Mod50to5000.xlsx')
mod_filtri = pd.read_excel('modfiltriaperti.xlsx')
xyz_acc = pd.read_excel('xyzacc50hz.xlsx')

# Anteprima dei dati caricati
print("MainISOfiltriaperti.xlsx")
display(main_iso.head())

print("Mod50to5000.xlsx")
display(mod_50_5000.head())

print("modfiltriaperti.xlsx")
display(mod_filtri.head())

print("xyzacc50hz.xlsx")
display(xyz_acc.head())

import numpy as np

# Funzione di analisi esplorativa rapida per ogni dataset
def quick_eda(df, name):
    print(f"\n Analisi preliminare: {name}")
    print("-" * 60)

    print("Shape:", df.shape)
    print("Tipi di dati:")
    print(df.dtypes)

    print("\n Valori nulli per colonna:")
    print(df.isnull().sum())

    print("\n Statistiche descrittive:")
    display(df.describe())

    print("\n Valori unici per ogni colonna:")
    for col in df.columns:
        unique_vals = df[col].nunique()
        print(f"  - {col}: {unique_vals} valori unici")

#  Applica la funzione a ciascun file
quick_eda(main_iso, "MainISOfiltriaperti.xlsx")
quick_eda(mod_50_5000, "Mod50to5000.xlsx")
quick_eda(mod_filtri, "modfiltriaperti.xlsx")
quick_eda(xyz_acc, "xyzacc50hz.xlsx")
```

```python
import matplotlib.pyplot as plt
n = 10000
df_plot = xyz_acc.head(n).copy()
df_plot['Time[s]'] = df_plot['Time[ms]'] / 1000000  # converte in secondi

# Plot asse X
plt.figure(figsize=(14, 4))
plt.plot(df_plot['Time[s]'], df_plot['X'], color='steelblue')
plt.title(f'Accelerazione Asse X  Primi {n} campioni')
plt.xlabel('Tempo (s)')
plt.ylabel('Accelerazione X (m/s)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot asse Y
plt.figure(figsize=(14, 4))
plt.plot(df_plot['Time[s]'], df_plot['Y'], color='seagreen')
plt.title(f'Accelerazione Asse Y  Primi {n} campioni')
plt.xlabel('Tempo (s)')
plt.ylabel('Accelerazione Y (m/s)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot asse Z
plt.figure(figsize=(14, 4))
plt.plot(df_plot['Time[s]'], df_plot['Z'], color='darkorange')
plt.title(f'Accelerazione Asse Z  Primi {n} campioni')
plt.xlabel('Tempo (s)')
plt.ylabel('Accelerazione Z (m/s)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#acceleration module
import numpy as np
import matplotlib.pyplot as plt

mod_no_filter = xyz_acc.copy()
mod_no_filter['Modulo'] = np.sqrt(
    mod_no_filter['X']**2 +
    mod_no_filter['Y']**2 +
    mod_no_filter['Z']**2
)

#time in sec
mod_no_filter['Time[s]'] = mod_no_filter['Time[ms]'] / 1000000

# plot
n = 5000
df_mod_plot = mod_no_filter.head(n)

plt.figure(figsize=(14, 6))
plt.plot(df_mod_plot['Time[s]'], df_mod_plot['Modulo'], color='purple', linewidth=1)
plt.title(f'Modulo accelerazione risultante (primi {n} campioni)', fontsize=16)
plt.xlabel('Tempo (s)', fontsize=14)
plt.ylabel('Modulo [m/s]', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# remove mean from each axis
acc_x = xyz_acc['X'].values - np.mean(xyz_acc['X'].values)
acc_y = xyz_acc['Y'].values - np.mean(xyz_acc['Y'].values)
acc_z = xyz_acc['Z'].values - np.mean(xyz_acc['Z'].values)

# define time between 2 samples
dt = 0.000000375  # 375 microsecondi

# integration to have velocity on each axis
vel_x = np.cumsum(acc_x) * dt
vel_y = np.cumsum(acc_y) * dt
vel_z = np.cumsum(acc_z) * dt

# velocity module (same formula as before)
vel_mod = np.sqrt(vel_x**2 + vel_y**2 + vel_z**2)

# add time and velocity module to a dataframe
mod_vel = pd.DataFrame({
    'Time[ms]': xyz_acc['Time[ms]'],
    'Time[s]': xyz_acc['Time[ms]'] / 1000000,
    'Modulo_velocit [m/s]': vel_mod
})

# plot
plt.figure(figsize=(14, 5))
plt.plot(mod_vel['Time[s]'][:5000], mod_vel['Modulo_velocit [m/s]'][:5000], color='teal')
plt.title('Modulo velocit risultante (pseudo-integrata)', fontsize=15)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocit [m/s]')
plt.grid(True)
plt.tight_layout()
plt.show()
```
